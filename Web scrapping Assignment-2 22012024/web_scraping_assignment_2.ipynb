Q1: Write a python program to scrape data for ‚ÄúData Analyst‚Äù Job position in ‚ÄúBangalore‚Äù location. You
have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10
jobs data.
This task will be done in following steps:
1. First get the webpage https://www.shine.com/
2. Enter ‚ÄúData Analyst‚Äù in ‚ÄúJob title, Skills‚Äù field and enter ‚ÄúBangalore‚Äù in ‚Äúenter the location‚Äù field.
3. Then click the searchbutton.
4. Then scrape the data for the first 10 jobs results you get.
5. Finally create a dataframe of the scraped data.
Note: All of the above steps have to be done in code. No step is to be done manually.

```python
# Step 1: Open the webpage
url = "https://www.shine.com/"
driver = webdriver.Chrome()  # You need to have Chrome WebDriver installed and its path configured
driver.get(url)
search=driver.find_element(By.XPATH,'//input[@class="input"]')
search.click()
time.sleep(5)
# Step 2: Enter search criteria and click search button
search_keyword = "Data Analyst"
search_location = "Bangalore"

job_title_field =driver.find_element(By.XPATH,'//input[@id="id_q"]')
job_title_field.clear()
job_title_field.send_keys(search_keyword)

location_field =driver.find_element(By.XPATH,'//input[@id="id_loc"]') 
location_field.clear()
location_field.send_keys(search_location)

search_button = driver.find_element(By.XPATH,'//button[@class=" btn btn-secondary undefined"]')
search_button.click()

# Wait for the page to load
time.sleep(5)

job_titles = []
job_locations = []
company_names = []
experience_required = []


job_listings = driver.find_elements(By.XPATH,'//h2[@itemprop="name"]/a')
for i in job_listings[:10]:
    title = i.text
    job_titles.append(title)

#print(job_titles)


job_Locations = driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"]')
for i in job_Locations[:10]:
    location = i.text
    job_locations.append(location)

#print(job_locations)

company_Names = driver.find_elements(By.XPATH,'//div[@class="jobCard_jobCard_cName__mYnow"]/span')
for i in company_Names[:10]:
    comapany = i.text
    company_names.append(comapany)

#print(company_names)

experience_Required = driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"]')
for i in experience_Required[:10]:
    experience = i.text
    experience_required.append(experience)

#print(experience_required)

# Step 4: Create a dataframe of the scraped data
data = ({
    'Job Title': job_titles,
    'Job Location': job_locations,
    'Company Name': company_names,
    'Experience Required': experience_required
})

df = pd.DataFrame(data)
# Display the dataframe
print(df)

# Close the browser
driver.quit()
```

Q9: Write a python program to display list of respected former Prime Ministers of India(i.e. Name, Born-Dead,
Term of office, Remarks) from https://www.jagranjosh.com/.
This task will be done in following steps:
1. First get the webpagehttps://www.jagranjosh.com/
2. Then You have to click on the GK option
3. Then click on the List of all Prime Ministers of India
4. Then scrap the mentioned data and make theDataFrame

Note:-nable to find  List of all Prime Ministers of India so adjusted code accordingly

```python
# Set up Selenium webdriver (make sure you have installed the appropriate webdriver for your browser)
driver = webdriver.Chrome()  # You need to have Chrome WebDriver installed and its path configured
driver.get("https://www.jagranjosh.com/")

# Step 1: Click on the "GK" option
gk_link = driver.find_element(By.XPATH,'//a[@href="https://www.jagranjosh.com/general-knowledge"]')
gk_link.click()
time.sleep(5)

# Step 2: Click on the "List of all Prime Ministers of India" link
prime_ministers_link = driver.find_element(By.XPATH,'//a[@href="https://www.jagranjosh.com/general-knowledge/list-of-finance-ministers-of-india-1473760031-1"]')
prime_ministers_link.click()
time.sleep(5)

# Step 3: Scrap the mentioned data
data = []

# Find the table containing the Prime Ministers data
table = driver.find_element(By.XPATH,'//div[@class="TableData"]')

# Extract data from the table
rows = table.find_elements(By.TAG_NAME, "tr")
for row in rows[1:]:
    cols = row.find_elements(By.TAG_NAME, "td")
    if len(cols) >= 3: 
        name = cols[0].text.strip()
        term_of_office = cols[2].text.strip()  
        data.append([name, term_of_office])

# Step 4: Create a DataFrame
df = pd.DataFrame(data, columns=["Name", "Term of Office"])

# Step 5: Display the DataFrame
print(df)

# Close the browser window when done
driver.quit()
```
